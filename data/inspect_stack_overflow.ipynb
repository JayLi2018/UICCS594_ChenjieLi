```python
df = pd.read_csv('so_countries_col_new.csv', index_col=None)
```

```python
grouped_df = df.groupby(['Country'])['ConvertedSalary'].mean().reset_index().sort_values(by=['ConvertedSalary'])
grouped_df = grouped_df.rename(columns={'ConvertedSalary': 'avgsalary'})
grouped_df['avgsalary'] = grouped_df['avgsalary'].round().astype(int)
display(grouped_df)
```
