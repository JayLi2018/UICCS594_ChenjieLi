```python

import sys
import os
import pandas as pd
import numpy as np
from causallearn.utils.GraphUtils import GraphUtils
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import io

sys.path.append(os.path.join(os.getcwd(), ".."))
```

```python
from synthetic_data_generation import gen_single_relation, gen_single_relation_with_hidden_confounder
```

```python
df = gen_single_relation()
```

```python
from causallearn.search.ConstraintBased.PC import pc
from causallearn.search.ConstraintBased.FCI import fci
```

```python
def pc_function(df):
    labels = [f'{col}' for i, col in enumerate(df.columns)]
    data = df.to_numpy()
    cg = pc(data)
    pyd = GraphUtils.to_pydot(cg.G, labels=labels)
    tmp_png = pyd.create_png(f="png")
    fp = io.BytesIO(tmp_png)
    img = mpimg.imread(fp, format='png')
    plt.figure(figsize=(8, 6))  
    plt.axis('off')
    plt.imshow(img)
    plt.show()
    
```

```python
pc_function(df=df)
```

```python
def fci_function(df):
    labels = [f'{col}' for i, col in enumerate(df.columns)]
    data = df.to_numpy()
    cg, edges = fci(data)
    # print(edges)
    pyd = GraphUtils.to_pydot(cg,labels=labels)
    tmp_png = pyd.create_png(f="png")
    fp = io.BytesIO(tmp_png)
    img = mpimg.imread(fp, format='png')
    plt.figure(figsize=(8, 6))  
    plt.axis('off')
    plt.imshow(img)
    plt.show()
```

```python
fci_function(df=df)
```

```python
for a_s in [0,0.1,0.3,0.5,0.7,0.9]:
    for b_s in [0,0.1,0.3,0.5,0.7,0.9]:
        df_with_latent_var = gen_single_relation_with_hidden_confounder(l_to_a=a_s, l_to_b=b_s, n_rows=1000000)
        print(f"results for a_s ={a_s}, b_s={b_s}")
        print(f"for pc")
        pc_function(df=df_with_latent_var)
        print(f"for fci")
        fci_function(df=df_with_latent_var)
        print('-------------------------------------\n')
```
